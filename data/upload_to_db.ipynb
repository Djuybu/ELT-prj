```python
# Bước 1: Chuẩn hóa lại DataFrame sản phẩm
product_df = product_df.rename(columns={
    'id': 'product_id',
    'brand': 'product_brand'
})
product_df = product_df[['product_id', 'product_name', 'price', 'description', 'img_link', 'product_brand']]
product_df = product_df.drop_duplicates(subset=['product_id'])

# Bước 2: Lưu product_df vào bảng sản phẩm
product_df.to_sql(
    name='product',        # tên bảng sản phẩm
    con=engine,
    if_exists='append',
    index=False,
    dtype={
        'product_id': Integer,
        'product_name': String(255),
        'price': REAL,
        'description': String(255),
        'img_link': String(255),
        'product_brand': String(255)
    }
)
```

```python
category_df = category_df.rename(columns={
        'category_id': 'id',
        'category_name': 'category_name'
    })

category_df.to_sql(
    name='category',        # tên bảng sản phẩm
    con=engine,
    if_exists='append',
    index=False,
    dtype={
        'id': Integer,
        'category_name': String
    }
)
```
